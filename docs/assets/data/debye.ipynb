```python
import numpy as np
import scipy.integrate as spi
```

```python
def integrando(u):
    return u**4 / (np.exp(u)-1)**2
```

```python
def integral(x):
    valor, erro =  spi.quad(integrando, 0, x)
    return valor
```

```python
def D(x):
    return 9 / x**3 * integral(x)
```

```python
def Cv(T, R=8.31451, thD=1849):
    x = thD
    return R * D(x)
```
